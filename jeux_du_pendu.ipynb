#### De quoi a-t-on besoin pour jouer au pendu:
- choisir un mot
- deviner une lettre
- verifier si la lettre est dans le mot
- mettre à jour les lettres trouvées
- ajouter un baton (retirer une vie)
- jouer jusqu'à ce qu'on ait plus de vie ou qu'on ait trouvé le mot


```python
import getpass
```

On veut ranger chaqune de ces fonctionalités dans une fonction qu'on pourra appeler quand on en aura besoin. 

Pour créer une nouvelle fonction, on utilise <font color='green'>def</font>:

```python
def choisir_un_mot():
    mot = getpass.getpass("Choisis le mot à deviner ")
    mot = mot.lower() # on met tout en minuscule parceque pour python a != A
    return mot

mot_secret= choisir_un_mot()
```
