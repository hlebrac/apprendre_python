#### De quoi a-t-on besoin pour jouer au pendu:
- choisir un mot
- deviner une lettre
- verifier si la lettre est dans le mot
- mettre  jour les lettres trouves
- ajouter un baton (retirer une vie)
- jouer jusqu' ce qu'on ait plus de vie ou qu'on ait trouv le mot


```python
import getpass
```

On veut ranger chaqune de ces fonctionalits dans une fonction qu'on pourra appeler quand on en aura besoin. 

Pour crer une nouvelle fonction, on utilise <font color='green'>def</font>:

```python
def choisir_un_mot():
    mot = getpass.getpass("Choisis le mot  deviner ")
    mot = mot.lower() # on met tout en minuscule parceque pour python a != A
    return mot

mot_secret= choisir_un_mot()
```
